```python
# standerd imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('mobile_detail.csv')
df = df.drop(columns='corpus')
df = df.drop(columns='imgURL')
df
```

```python
df.isnull().sum()
df = df.dropna()
```

```python
# Build ML Models
X = df.drop('price', axis = 1)
y = df['price']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.20)
```

```python
# Build ML model
from sklearn.ensemble import RandomForestRegressor
phon_model = RandomForestRegressor()
phon_model.fit(X_train, y_train)
phon_model.score(X_test, y_test)
```

```python
# encoding
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
feature_cat = ["brand","mobile_name","color", "system","system","processor_type","resolution"]
featured = OneHotEncoder()
trans = ColumnTransformer([('featured_final',featured,feature_cat)],
                         remainder="passthrough")
transformed_X = trans.fit_transform(X)

pd.DataFrame(transformed_X)
```
