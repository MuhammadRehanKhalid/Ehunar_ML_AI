## Setting Up Features

```python
# prepare data
x = [1,2,3,4,5,6,7]
y = [12,34,67,56,78,34,19]
# Create Canvas
home, room = plt.subplots()
# Create plot
room.plot(x,y)
# Coustomization
room.set(title = 'Muhammad Rehan Khalid',
        xlabel = 'X-Axis',
        ylabel = 'Y-Axis')
plt.show()
```

## Plot Multiple graph in one figure

```python
# prepare data
x = [1,2,3,4,5,6,7]
y = [12,34,67,56,78,34,19]
# Create Canvas for multiple rooms
# need fig size

home, ((room1, room2),(room3, room4)) = plt.subplots(nrows=2, ncols=2, figsize = (10,10))
# Create plot
room1.plot(x,y)
room2.plot(x,y)
room3.plot(x,y)
room4.plot(x,y)
# Coustomization
room.set(title = 'Muhammad Rehan Khalid',
        xlabel = 'X-Axis',
        ylabel = 'Y-Axis')
plt.show()
```
