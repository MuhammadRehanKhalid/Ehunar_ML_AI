## Setting Up Features

```python
# prepare data
x = [1,2,3,4,5,6,7]
y = [12,34,67,56,78,34,19]
# Create Canvas
home, room = plt.subplots()
# Create plot
room.plot(x,y)
# Coustomization
room.set(title = 'Muhammad Rehan Khalid',
        xlabel = 'X-Axis',
        ylabel = 'Y-Axis')
plt.show()
```

## Plot Multiple graph in one figure

```python
# prepare data
x = [1,2,3,4,5,6,7]
y = [12,34,67,56,78,34,19]
# Create Canvas for multiple rooms
# need fig size

home, ((room1, room2),(room3, room4)) = plt.subplots(nrows=2, ncols=2, figsize = (10,10))
# Create plot
room1.plot(x,y)
room2.plot(x,y)
room3.plot(x,y)
room4.plot(x,y)
# Coustomization
room.set(title = 'Muhammad Rehan Khalid',
        xlabel = 'X-Axis',
        ylabel = 'Y-Axis')
plt.show()
```

## Different Kinds of Basic Plots Most Used

```python
# Scatter plot
fig, sub_plot = plt.subplots()
sub_plot.scatter(x,y)
plt.show()
```

```python
# barplot 
fig, sub_plot = plt.subplots()
sub_plot.bar(x,y)
plt.show()
```

```python
shop = {'Item_1': 20,
       'Item _2': 32,
       'Item_3' : 44}
fig, sub_plot = plt.subplots()
sub_plot.bar(shop.keys(),shop.values());
plt.show()
```

```python
# for horizontal bar plot covert them to list
shop = {'Item_1': 20,
       'Item _2': 32,
       'Item_3' : 44}
fig, sub_plot = plt.subplots()
sub_plot.barh(list(shop.keys()),list(shop.values()));
plt.show()
```
