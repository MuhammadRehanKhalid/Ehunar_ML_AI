```python
# R^2, thumb rule is to maximize the R2 value
np.random.seed(4)
df_1 = pd.read_csv('housing.csv')
df_1
```

```python
# create data
np.random.seed(3)
X = df_1.drop('target', axis = 1)
y = df_1['target']
# Model split test and train
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.20)
# instantiate the Ridge Model
reg_model = RandomForestRegressor()
reg_model.fit(X_train, y_train)
reg_model.score(X_test, y_test)
```

# mean absolute error

```python
# mean absolute error
# MAE = the average of th absolute (plus) difference between predicted and actual
from sklearn.metrics import mean_absolute_error
y_pred = reg_model.predict(X_test)
MAE = mean_absolute_error(y_test, y_pred)

MAE
```

```python
# Converting into data
df = pd.DataFrame(data =  {'actual_values': y_test,
                          'predicted_Values': y_pred})

df['difference'] = df['actual_values'] - df['predicted_Values']
df
```

# Mean Square error

```python
# take sqaure of them, then add upp, take mean
```

```python
from sklearn.metrics import mean_squared_error
y_pred = reg_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
# mannualy 
mse_mannual = np.square(df["difference"])
mse_mannual.mean()
```
